```python
## heatmeap to see the correlation between features.
# Generate a mask for the upper triangle (taken from seaborn example gallery)
import numpy as np
mask = np.zeros_like(df.corr(), dtype=bool)
mask[np.triu_indices_from(mask)] = True
sns.set_style('whitegrid')
plt.subplots(figsize = (15,12))
sns.heatmap(df.corr(),
annot=True,
mask = mask,
cmap = 'Blues',
linewidths=.9,
linecolor='white',
fmt='.2g',
center = 0,
square=True)
plt.title("Correlations Among Features", y = 1.03,fontsize = 20, pad = 40);
```

<h3 style ="color:red"> Correlated Features: </h3>
<ul>
    <li style ="color:purple; list-style-type:upper-roman">cnt and instant, yr, temp atemp casual registered are correlated features</li>

    <li style ="color:purple; list-style-type:upper-roman"> casual and, temp, atemp, casual correlated features</li>
